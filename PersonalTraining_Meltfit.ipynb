```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the metrics
metrics = ['10-Meter Sprint', 'T-Test', '1RM (One-Rep Max) Bench Press', '1RM (One-Rep Max) Squat', 'VO2 Max Test', 'Shooting Accuracy']
num_metrics = len(metrics)

# Define the standards and current status
standards = [1.6, 10, 1.4, 1.1, 55, 90]
current_status = [1.7, 11, 1.2, 0.9, 50, 80]
```

```python
# Number of metrics
num_metrics = len(metrics)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_metrics, endpoint=False).tolist()
angles += angles[:1]

# Repeat the first value to close the circle
standards += standards[:1]
current_status += current_status[:1]

# Determine the max value for the y-axis labels
max_value = max(max(standards), max(current_status))

# Define the y-labels based on the max value
num_ticks = 5
yticks = np.linspace(0, max_value, num_ticks + 1)

# Plotting
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axe per variable and add labels
plt.xticks(angles[:-1], metrics)

# Draw y-labels
ax.set_rlabel_position(30)
plt.yticks(yticks[:-1], [str(int(y)) for y in yticks[:-1]], color="grey", size=7)
plt.ylim(0, max_value)

# Plot data
ax.plot(angles, standards, linewidth=1, linestyle='solid', label='Standard')
ax.fill(angles, standards, 'b', alpha=0.1)

ax.plot(angles, current_status, linewidth=1, linestyle='solid', label='Current Status')
ax.fill(angles, current_status, 'r', alpha=0.1)

# Add a legend
plt.legend(loc='upper right', bbox_to_anchor=(0.1, 0.1))

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the metrics, standards, and current status with units
metrics = ['10-Meter Sprint (s)', 'T-Test (s)', '1RM Bench Press (kg)', '1RM Squat (kg)', 'VO2 Max Test (ml/kg/min)', 'Shooting Accuracy (%)']
standards = [1.6, 10, 1.4, 1.1, 55, 90]
current_status = [1.7, 11, 1.2, 0.9, 50, 95]

# Define whether lower or higher is better for each metric
lower_better = [True, True, False, False, False, False]

# Calculate deviations (positive for good, negative for bad based on the criteria)
deviations = [
    (current - standard) if not is_lower_better else (standard - current)
    for current, standard, is_lower_better in zip(current_status, standards, lower_better)
]

# Flag metrics as 'green' or 'alert' based on deviation from standard
flags = ['green' if d >= 0 else 'alert' for d in deviations]

# Convert flags to numerical values
flag_values = [1 if flag == 'green' else 0 for flag in flags]

# Create a data matrix with the deviations and flag row
data = np.vstack([standards, current_status, deviations, flag_values])

# Define colors based on deviations
deviation_colors = ['#f7f7f7', '#f7f7f7']  # Light gray for standard and current status
for d in deviations:
    if d < 0:
        deviation_colors.append('#fee0d2')  # Light red for bad deviations
    else:
        deviation_colors.append('#d9f0d3')  # Light green for good deviations

# Create a custom color map for deviations
deviation_cmap = sns.color_palette(deviation_colors)

# Convert flag values to corresponding colors
flag_colors = ['#d9f0d3' if flag == 1 else '#fee0d2' for flag in flag_values]

# Add the flag colors to the deviation color map
deviation_cmap.extend(sns.color_palette(flag_colors))

# Create the heatmap
plt.figure(figsize=(14, 4))  # Adjusting the height for a better visual
sns.heatmap(data, annot=True, fmt=".1f", cmap=deviation_cmap, xticklabels=metrics, yticklabels=['Standard', 'Current Status', 'Deviation', 'Flag'], cbar=False, linewidths=0.5, linecolor='gray')
plt.title('Performance Metrics Comparison', fontsize=18, fontweight='bold')
plt.xlabel('Metrics', fontsize=14, fontweight='bold')
plt.yticks(rotation=0)
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define the metrics, standards, and current status with units
metrics = ['10-Meter Sprint (s)', 'T-Test (s)', '1RM Bench Press (kg)', '1RM Squat (kg)', 'VO2 Max Test (ml/kg/min)', 'Shooting Accuracy (%)']
standards = [1.6, 10, 1.4, 1.1, 55, 90]
current_status = [1.7, 11, 1.2, 0.9, 50, 80]

# Define whether lower or higher is better for each metric
lower_better = [True, True, False, False, False, False]

# Calculate deviations (positive for good, negative for bad based on the criteria)
deviations = [
    (current - standard) if not is_lower_better else (standard - current)
    for current, standard, is_lower_better in zip(current_status, standards, lower_better)
]

# Flag metrics as 1 for 'within standard' or 0 for 'deviation from standard'
flags = [1 if d >= 0 else 0 for d in deviations]

# Create a data matrix with the deviations and flag row
data = np.array([standards, current_status, deviations, flags])

# Define colors based on deviations and flags
colors = ['#f7f7f7', '#f7f7f7']  # Light gray for standard and current status
for d, flag in zip(deviations, flags):
    if d < 0:
        colors.append('#fee0d2') if flag == 0 else colors.append('#f7f7f7')  # Light red for deviation from standard
    else:
        colors.append('#d9f0d3') if flag == 1 else colors.append('#f7f7f7')  # Light green for within standard

# Create a custom color map
cmap = sns.color_palette(colors)

# Create the heatmap
plt.figure(figsize=(14, 5))  # Adjusting the height for a better visual
sns.heatmap(data, annot=True, fmt=".1f", cmap=cmap, xticklabels=metrics, yticklabels=['Standard', 'Current Status', 'Deviation'], cbar=False, linewidths=0.5, linecolor='gray')
plt.title('Performance Metrics Comparison', fontsize=18, fontweight='bold')
plt.xlabel('Metrics', fontsize=14, fontweight='bold')
plt.yticks(rotation=0)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the weeks (assuming 6 weeks for illustration)
weeks = np.arange(1, 7)

# Define your current university ranking and goal
current_ranking = 150
goal_ranking = 100

# Generate a trend for your university ranking (e.g., assuming improvement each week)
# Here, we are using a simple linear trend, decreasing by 10 each week
ranking_trend = current_ranking - np.arange(6) * 10

# Plot the data
plt.figure(figsize=(10, 6))

# Plot your current university ranking trend
plt.plot(weeks, ranking_trend, marker='o', label='Current University Ranking', color='blue')

# Plot the goal university ranking
plt.hlines(goal_ranking, weeks[0], weeks[-1], linestyles='--', colors='red', label='Goal University Ranking')

# Add labels and title
plt.title('Weekly University Ranking Progress', fontsize=18, fontweight='bold')
plt.xlabel('Week', fontsize=14, fontweight='bold')
plt.ylabel('University Ranking', fontsize=14, fontweight='bold')

# Invert the y-axis to represent lower values as higher positions on the plot
plt.gca().invert_yaxis()

# Add legend
plt.legend()

# Show grid
plt.grid(True)

# Show plot
plt.show()
```
