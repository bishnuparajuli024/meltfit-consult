```python
import gspread
from google.colab import auth
from google.auth import default
import pandas as pd
import folium
from folium.plugins import HeatMap
import re

# Authenticate and create the PyDrive client.
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Open the Google Sheet
sheet = gc.open_by_key('1YHZR0BKjcFnrJ2Q-6Bbps0CJWcv8ZxzCieA0yAlxlDw')
worksheet = sheet.get_worksheet(0)

# Extract data from Google Sheet
data = worksheet.get_all_values()
df = pd.DataFrame(data[1:], columns=data[0])

# Extract latitude and longitude from Google Maps link
def extract_lat_long(url):
    match = re.search(r'@(-?\d+\.\d+),(-?\d+\.\d+)', url)
    if match:
        return float(match.group(1)), float(match.group(2))
    return None, None

df['Latitude'], df['Longitude'] = zip(*df['Map URL'].apply(extract_lat_long))

# Drop rows with missing lat/long
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Create a map centered around the mean coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

# Add heatmap layer
HeatMap(data=df[['Latitude', 'Longitude']].values.tolist(), radius=15, blur=10, max_zoom=1).add_to(m)

# Save the map to an HTML file
m.save('fitness_centers_heatmap.html')

# Display the map
m
```

```python
import gspread
from google.colab import auth
from google.auth import default
import pandas as pd
import folium
from folium.plugins import HeatMap
from geopy.geocoders import Nominatim

# Authenticate and create the PyDrive client.
auth.authenticate_user()
creds, _ = default()
gc = gspread.authorize(creds)

# Open the Google Sheet
sheet = gc.open_by_key('1YHZR0BKjcFnrJ2Q-6Bbps0CJWcv8ZxzCieA0yAlxlDw')
worksheet = sheet.get_worksheet(0)

# Extract data from Google Sheet
data = worksheet.get_all_values()
df = pd.DataFrame(data[1:], columns=data[0])

# Extract latitude and longitude from 'Location ID' column
def extract_lat_long(location_id):
    try:
        lat, lon = location_id.split(',')
        return float(lat.strip()), float(lon.strip())
    except Exception as e:
        print(f"Error extracting lat/long from '{location_id}': {e}")
        return None, None

df['Latitude'], df['Longitude'] = zip(*df['Location ID'].apply(extract_lat_long))

# Drop rows with missing lat/long
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Reverse geocode to get place names
geolocator = Nominatim(user_agent="gym_heatmap")
df['Place Name'] = df.apply(lambda row: geolocator.reverse((row['Latitude'], row['Longitude'])).address, axis=1)

# Create a map centered around the mean coordinates
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
m = folium.Map(location=map_center, zoom_start=12)

# Add markers with place names
for index, row in df.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Place Name']).add_to(m)

# Add heatmap layer with adjusted parameters for clarity
HeatMap(data=df[['Latitude', 'Longitude']].values.tolist(), radius=15, blur=10, max_zoom=1).add_to(m)

# Specify the viewport to zoom into the main heat areas
m.fit_bounds([[df['Latitude'].min(), df['Longitude'].min()], [df['Latitude'].max(), df['Longitude'].max()]])

# Save the map as a JPEG image
m.save('gym_partners_heatmap.jpg')

# Display the map
m
```
